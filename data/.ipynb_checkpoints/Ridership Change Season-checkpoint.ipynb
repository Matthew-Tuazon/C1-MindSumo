```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import collections
from collections import Counter #unique counter for arrays
#types of passes during season
springTrips = []
summerTrips = []
fallTrips = []
winterTrips = []

#for trip durations during season
springDuration = []
summerDuration = []
fallDuration = []
winterDuration = []

#for popular BikeID during season
springBike = []
summerBike = []
fallBike = []
winterBike = []
```

```python
%matplotlib inline
df = pd.read_csv('metro-bike-share-trip-data.csv')
```

```python
#Replaces spaces with _ and get rid of parenthesis 
df.columns = df.columns.str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
df["Trip_ID"].astype(int)
df["Duration"].astype(int)
df["Start_Time"].astype(str)
df["End_Time"].astype(str)
df["Starting_Station_ID"].astype(float)
df["Starting_Station_Latitude"].astype(float)
df["Starting_Station_Longitude"].astype(float)
df["Ending_Station_ID"].astype(float)
df["Ending_Station_Latitude"].astype(float)
df["Ending_Station_Longitude"].astype(float)
df["Bike_ID"].astype(float)
df["Trip_Route_Category"].astype(str)
df["Passholder_Type"].astype(str)
df["Starting_Lat-Long"].astype(str)
df["Ending_Lat-Long"].astype(str)
```

```python
#converts formatting of Start_Time/End_Time data to dateTime
df["Start_Time"] = pd.to_datetime(df['Start_Time'])
df["End_Time"] = pd.to_datetime(df['End_Time'])
```

```python
#checks season using the dateTime
#used season time frames based on: https://www.timeanddate.com/calendar/aboutseasons.html
def season(datetime):
    if 3 <= datetime.month and datetime.month <= 5:
        return "Spring"
    elif 6 <= datetime.month and datetime.month<= 8:
        return "Summer"
    elif 9 <= datetime.month and datetime.month<= 11:
        return "Fall/Autumn"
    elif 12 == datetime.month or datetime.month <= 2:
        return "Winter"

#Adds seasons to df chart
seasons = pd.Series(map(season,df["Start_Time"]))
df["Season"] = seasons
```

```python
Spring = df[df.Season == 'Spring']
Summer = df[df.Season == 'Summer']
Fall = df[df.Season == 'Fall/Autumn']
Winter = df[df.Season == 'Winter']
```

```python
#Spring #13906 Trips in Spring
#Summer #35573 Trips in the Summer
#Fall #52455 Trips in the Fall
#Winter #30493 Trips in Winter
```

```python
#For Loop to add different types of pass used for each season.
for index, row in df.iterrows():
    passUsed = row["Passholder_Type"]
    season = row["Season"]
    if season == "Spring":
        springTrips.append(passUsed)
    elif season == "Summer":
        summerTrips.append(passUsed)
    elif season == "Fall/Autumn":
        fallTrips.append(passUsed)
    elif season == "Winter":
        winterTrips.append(passUsed)
```

```python
#Different types of passes used per season
print ("Summer Trips Count:")
print ("Monthly passes count - {}".format(summerTrips.count("Monthly Pass")))
print ("Flex passes count - {}".format(summerTrips.count("Flex Pass")))
print ("Walk-up passes count - {}".format(summerTrips.count("Walk-up")))
print ("Staff Annual passes count - {}\n".format(summerTrips.count("Staff Annual")))
print ("Spring Trips Count:")
print ("Monthly passes count - {}".format(springTrips.count("Monthly Pass")))
print ("Flex passes count - {}".format(springTrips.count("Flex Pass")))
print ("Walk-up passes count - {}".format(springTrips.count("Walk-up")))
print ("Staff Annual passes count - {}\n".format(springTrips.count("Staff Annual")))
print ("Fall/Autumn Trips Count:")
print ("Monthly passes count - {}".format(fallTrips.count("Monthly Pass")))
print ("Flex passes count - {}".format(fallTrips.count("Flex Pass")))
print ("Walk-up passes count - {}".format(fallTrips.count("Walk-up")))
print ("Staff Annual passes count - {}\n".format(fallTrips.count("Staff Annual")))
print ("Winter Trips Count:")
print ("Monthly passes count - {}".format(winterTrips.count("Monthly Pass")))
print ("Flex passes count - {}".format(winterTrips.count("Flex Pass")))
print ("Walk-up passes count - {}".format(winterTrips.count("Walk-up")))
print ("Staff Annual passes count - {}\n".format(winterTrips.count("Staff Annual")))
```

```python
#For Loop to track the average duration per season 
for index, row in df.iterrows():
    tripDuration = row["Duration"]
    season = row["Season"]
    if season == "Spring":
        springDuration.append(tripDuration)
    elif season == "Summer":
        summerDuration.append(tripDuration)
    elif season == "Fall/Autumn":
        fallDuration.append(tripDuration)
    elif season == "Winter":
        winterDuration.append(tripDuration)
```

```python
#Average duration of trips per season
print ("Average spring trip duration - {}".format(sum(springDuration)/len(springDuration)/60))
print ("Average summer trip duration - {}".format(sum(summerDuration)/len(summerDuration)/60))
print ("Average fall trip duration - {}".format(sum(fallDuration)/len(fallDuration)/60))
print ("Average winter trip duration - {}".format(sum(winterDuration)/len(winterDuration)/60))
```

```python
#For Loop to calculate most used Bikes
for index, row in df.iterrows():
    tripBikeID = row["Bike_ID"]
    season = row["Season"]
    #print (tripBikeID)
    if season == "Spring":
        springBike.append(tripBikeID)
    elif season == "Summer":
        summerBike.append(tripBikeID)
    elif season == "Fall/Autumn":
        fallBike.append(tripBikeID)
    elif season == "Winter":
        winterBike.append(tripBikeID)
```

```python
#most popular bike usage in spring, bikeID 5755
bikeSpringUsage = Counter(springBike)
bikeSpringUsage.most_common()
```

```python
#most popular bike usage in summer, bikeID 6487
bikeSummerUsage = Counter(summerBike)
bikeSummerUsage.most_common()
```

```python
#most popular bike usage in fall, bikeID 6527
bikeFallUsage = Counter(fallBike)
bikeFallUsage.most_common()
```

```python
#most popular bike usage in winter, bikeID 4727
bikeWinterUsage = Counter(winterBike)
bikeWinterUsage.most_common()
```
