```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import collections
from collections import Counter #unique counter for arrays

#Locations
downTownLA = []
portLA = []
venice = []
pasadena = []
```

```python
%matplotlib inline
df = pd.read_csv('metro-bike-share-trip-data.csv')
```

```python
#Replaces spaces with _ and get rid of parenthesis 
df.columns = df.columns.str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
df["Trip_ID"].astype(int)
df["Duration"].astype(int)
df["Start_Time"].astype(str)
df["End_Time"].astype(str)
df["Starting_Station_ID"].astype(float)
df["Starting_Station_Latitude"].astype(float)
df["Starting_Station_Longitude"].astype(float)
df["Ending_Station_ID"].astype(float)
df["Ending_Station_Latitude"].astype(float)
df["Ending_Station_Longitude"].astype(float)
df["Bike_ID"].astype(float)
df["Trip_Route_Category"].astype(str)
df["Passholder_Type"].astype(str)
df["Starting_Lat-Long"].astype(str)
df["Ending_Lat-Long"].astype(str)
```

```python
#Checks Location using the Starting Station ID
#Found Locations based on Starting Station ID's found on: https://bikeshare.metro.net/about/data/
def location(Starting_Station_ID):
    if 3005 <= Starting_Station_ID and Starting_Station_ID <= 3011 or 3014 <= Starting_Station_ID and Starting_Station_ID <= 4108 or Starting_Station_ID == 4194 or 4220 <= Starting_Station_ID and Starting_Station_ID <= 4273:
        return "DowntownLA"
    elif Starting_Station_ID == 3013 or 4125 <= Starting_Station_ID and Starting_Station_ID <= 4136:
        return "PortLA"
    elif 4202 <= Starting_Station_ID and Starting_Station_ID <= 4217:
        return "Venice"
    elif 4138 <= Starting_Station_ID and Starting_Station_ID <= 4183:
        return "Pasadena"

#Adds locations to df chart
locations = pd.Series(map(location,df["Starting_Station_ID"]))
df["Location"] = locations
```

```python
df["Starting_Station_ID"].count()
```

```python
for index, row in df.iterrows():
    location = row["Location"]
    if location == "DowntownLA":
        downTownLA.append(location)
    elif location == "PortLA":
        portLA.append(location)
    elif location == "Venice":
        venice.append(location)
    elif location == "Pasadena":
        pasadena.append(location)
```

```python
print ("Downtown LA Total Trips - {}".format(downTownLA.count("DowntownLA")))
print ("Port LA Total Trips - {}".format(portLA.count("PortLA")))
print ("Venice Total Trips - {}".format(venice.count("Venice")))
print ("Pasadena Total Trips - {}".format(pasadena.count("Pasadena")))
```
